## Total bill skewness

```python
sns.histplot(data=df,x="total_bill",bins=10)
```

```python
tb_mean = df["total_bill"].mean()
tb_med = df["total_bill"].median()
if tb_mean > tb_med:
    print("Mean > Median - positive skewness")
elif tb_mean < tb_med:
     print("Mean > Median - positive skewness")
else:
     print("Mean ~= Median - uniform skewness")
```
